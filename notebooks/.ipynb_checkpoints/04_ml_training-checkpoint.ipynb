```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.metrics import (accuracy_score, f1_score, classification_report,
                             confusion_matrix, mean_absolute_error,
                             mean_squared_error, r2_score)
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Brand colours
NAVY = '#0C447C'
BLUE = '#185FA5'
RED = '#E24B4A'
OFF_WHITE = '#F1EFE8'

print("✅ ML libraries loaded successfully")
```

```python
# Load cleaned datasets
df_main = pd.read_csv('../data/cleaned/cleaned_main.csv')
df_national = pd.read_csv('../data/cleaned/cleaned_national.csv')

# Filter provinces only
df_provinces = df_main[df_main['Province'] != 'Pakistan (National)'].copy()

print("✅ Data loaded")
print(f"Shape: {df_provinces.shape}")
print(f"\nColumns: {df_provinces.columns.tolist()}")
print(f"\nSeverity Label distribution:")
print(df_provinces['Severity_Label'].value_counts())
```

```python
# Encode categorical features
le_province = LabelEncoder()
le_severity = LabelEncoder()

df_ml = df_provinces.copy()

# Encode Province
df_ml['Province_Encoded'] = le_province.fit_transform(df_ml['Province'])

# Encode Severity Label (target for classifier)
df_ml['Severity_Encoded'] = le_severity.fit_transform(df_ml['Severity_Label'])

print("Severity encoding map:")
for i, label in enumerate(le_severity.classes_):
    print(f"  {i} = {label}")

# Define features for classifier
classifier_features = [
    'Province_Encoded',
    'Year_Numeric',
    'Total_Accidents',
    'Fatal_Accidents',
    'Non_Fatal_Accidents',
    'Total_Vehicles_Involved',
    'Fatality_Rate',
    'Injury_Rate',
    'Fatal_Accident_Ratio',
    'Vehicles_Per_Accident',
]

# Define features for regressor
regressor_features = [
    'Province_Encoded',
    'Year_Numeric',
    'Total_Accidents',
    'Fatal_Accidents',
    'Non_Fatal_Accidents',
    'Total_Vehicles_Involved',
    'Fatality_Rate',
    'Injury_Rate',
    'Severity_Index',
]

print(f"\n✅ Features prepared")
print(f"Classifier features: {len(classifier_features)}")
print(f"Regressor features:  {len(regressor_features)}")
```

```python
# --- CLASSIFIER --- Predict Severity Label
X_clf = df_ml[classifier_features]
y_clf = df_ml['Severity_Encoded']

# --- REGRESSOR --- Predict Killed
X_reg = df_ml[regressor_features]
y_reg = df_ml['Killed']

# Scale features
scaler_clf = StandardScaler()
scaler_reg = StandardScaler()

X_clf_scaled = scaler_clf.fit_transform(X_clf)
X_reg_scaled = scaler_reg.fit_transform(X_reg)

# Split — stratified for classifier
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(
    X_clf_scaled, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)

X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg_scaled, y_reg, test_size=0.2, random_state=42
)

print("✅ Train/Test split done")
print(f"Classifier — Train: {X_clf_train.shape} | Test: {X_clf_test.shape}")
print(f"Regressor  — Train: {X_reg_train.shape} | Test: {X_reg_test.shape}")
print(f"\nClass distribution in test set:")
for enc, label in zip(range(3), le_severity.classes_):
    count = (y_clf_test == enc).sum()
    print(f"  {label}: {count}")
```

```python
# --- BASELINE CLASSIFIER --- Logistic Regression
lr_clf = LogisticRegression(random_state=42, max_iter=1000)
lr_clf.fit(X_clf_train, y_clf_train)
y_clf_pred_lr = lr_clf.predict(X_clf_test)

lr_acc = accuracy_score(y_clf_test, y_clf_pred_lr)
lr_f1 = f1_score(y_clf_test, y_clf_pred_lr, average='weighted')

print("=" * 50)
print("BASELINE — Logistic Regression Classifier")
print("=" * 50)
print(f"Accuracy: {lr_acc:.4f}")
print(f"F1 Score: {lr_f1:.4f}")

# --- BASELINE REGRESSOR --- Linear Regression
lr_reg = LinearRegression()
lr_reg.fit(X_reg_train, y_reg_train)
y_reg_pred_lr = lr_reg.predict(X_reg_test)

lr_mae = mean_absolute_error(y_reg_test, y_reg_pred_lr)
lr_rmse = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred_lr))
lr_r2 = r2_score(y_reg_test, y_reg_pred_lr)

print("\n" + "=" * 50)
print("BASELINE — Linear Regression Regressor")
print("=" * 50)
print(f"MAE:  {lr_mae:.2f}")
print(f"RMSE: {lr_rmse:.2f}")
print(f"R²:   {lr_r2:.4f}")
```

```python
# --- RANDOM FOREST CLASSIFIER ---
rf_clf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced'
)
rf_clf.fit(X_clf_train, y_clf_train)
y_clf_pred_rf = rf_clf.predict(X_clf_test)

rf_acc = accuracy_score(y_clf_test, y_clf_pred_rf)
rf_f1 = f1_score(y_clf_test, y_clf_pred_rf, average='weighted')

print("=" * 50)
print("Random Forest Classifier Results")
print("=" * 50)
print(f"Accuracy: {rf_acc:.4f}")
print(f"F1 Score: {rf_f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_clf_test, y_clf_pred_rf,
      target_names=le_severity.classes_))

# Confusion Matrix
cm = confusion_matrix(y_clf_test, y_clf_pred_rf)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le_severity.classes_,
            yticklabels=le_severity.classes__)
plt.title('Random Forest — Confusion Matrix', color=NAVY)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

print("✅ Random Forest Classifier done")
```

```python
# Fixed confusion matrix — correct classes_ (single underscore)
cm = confusion_matrix(y_clf_test, y_clf_pred_rf)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le_severity.classes_,
            yticklabels=le_severity.classes_)
plt.title('Random Forest — Confusion Matrix', color=NAVY)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
print("✅ Confusion matrix done")
```

```python
# --- GRADIENT BOOSTING REGRESSOR ---
gb_reg = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
gb_reg.fit(X_reg_train, y_reg_train)
y_reg_pred_gb = gb_reg.predict(X_reg_test)

gb_mae = mean_absolute_error(y_reg_test, y_reg_pred_gb)
gb_rmse = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred_gb))
gb_r2 = r2_score(y_reg_test, y_reg_pred_gb)

print("=" * 50)
print("Gradient Boosting Regressor Results")
print("=" * 50)
print(f"MAE:  {gb_mae:.2f}")
print(f"RMSE: {gb_rmse:.2f}")
print(f"R²:   {gb_r2:.4f}")

# Plot actual vs predicted
plt.figure(figsize=(7, 4))
plt.scatter(y_reg_test, y_reg_pred_gb, color=BLUE, alpha=0.7, edgecolors=NAVY)
plt.plot([y_reg_test.min(), y_reg_test.max()],
         [y_reg_test.min(), y_reg_test.max()],
         color=RED, linewidth=2, linestyle='--', label='Perfect Fit')
plt.xlabel('Actual Killed')
plt.ylabel('Predicted Killed')
plt.title('Gradient Boosting — Actual vs Predicted', color=NAVY)
plt.legend()
plt.tight_layout()
plt.show()

print("✅ Gradient Boosting Regressor done")
```

```python
# --- FEATURE IMPORTANCES ---

# Classifier feature importances
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Random Forest feature importance
rf_importances = pd.Series(
    rf_clf.feature_importances_,
    index=classifier_features
).sort_values(ascending=True)

rf_importances.plot(kind='barh', ax=axes[0], color=BLUE)
axes[0].set_title('Random Forest — Feature Importances', color=NAVY)
axes[0].set_xlabel('Importance Score')

# Gradient Boosting feature importance
gb_importances = pd.Series(
    gb_reg.feature_importances_,
    index=regressor_features
).sort_values(ascending=True)

gb_importances.plot(kind='barh', ax=axes[1], color=RED)
axes[1].set_title('Gradient Boosting — Feature Importances', color=NAVY)
axes[1].set_xlabel('Importance Score')

plt.tight_layout()
plt.show()

print("✅ Feature importances plotted")
```

```python
import os

# Create models folder if not exists
os.makedirs('../models', exist_ok=True)

# Save Random Forest Classifier
joblib.dump(rf_clf, '../models/severity_classifier_rf.joblib')
joblib.dump(scaler_clf, '../models/scaler_classifier.joblib')
joblib.dump(le_severity, '../models/label_encoder_severity.joblib')
joblib.dump(le_province, '../models/label_encoder_province.joblib')

# Save Gradient Boosting Regressor
joblib.dump(gb_reg, '../models/fatality_regressor_gb.joblib')
joblib.dump(scaler_reg, '../models/scaler_regressor.joblib')

print("✅ All models saved!")
print("\nFiles in models/ folder:")
for f in os.listdir('../models'):
    size = os.path.getsize(f'../models/{f}')
    print(f"  {f} — {size/1024:.1f} KB")
```
