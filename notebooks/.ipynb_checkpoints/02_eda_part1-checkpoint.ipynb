```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# RoadGuard Pakistan brand colours
NAVY = '#0C447C'
BLUE = '#185FA5'
SKY = '#378ADD'
RED = '#E24B4A'
DEEP_RED = '#A32D2D'
OFF_WHITE = '#F1EFE8'

print("✅ Libraries and brand colours loaded")
```

```python
# Load all cleaned datasets
df_main = pd.read_csv('../data/cleaned/cleaned_main.csv')
df_national = pd.read_csv('../data/cleaned/cleaned_national.csv')
df_punjab = pd.read_csv('../data/cleaned/cleaned_punjab.csv')

# Filter out national aggregate row for province-level charts
df_provinces = df_main[df_main['Province'] != 'Pakistan (National)'].copy()
df_national_only = df_main[df_main['Province'] == 'Pakistan (National)'].copy()

print("✅ Datasets loaded")
print(f"Main:      {df_main.shape}")
print(f"Provinces: {df_provinces.shape}")
print(f"National:  {df_national.shape}")
print(f"Punjab:    {df_punjab.shape}")
```

```python
# Chart 01 — National Accident Trend Over Years
df_trend = df_national_only.sort_values('Year_Numeric')

fig = px.line(
    df_trend,
    x='Year_Numeric',
    y='Total_Accidents',
    title='🚗 Road Accident Trend in Pakistan (2006–2023)',
    markers=True,
    color_discrete_sequence=[BLUE]
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Year',
    yaxis_title='Total Accidents',
    xaxis=dict(tickmode='linear', tick0=2006, dtick=2),
    hovermode='x unified'
)

fig.update_traces(line=dict(width=3), marker=dict(size=8, color=RED))
fig.show()

print("✅ Chart 01 done")
```

```python
# Chart 02 — Total Accidents by Province
df_prov_total = df_provinces.groupby('Province')['Total_Accidents'].sum().reset_index()
df_prov_total = df_prov_total.sort_values('Total_Accidents', ascending=True)

fig = px.bar(
    df_prov_total,
    x='Total_Accidents',
    y='Province',
    orientation='h',
    title='🏙️ Total Accidents by Province (All Years)',
    color='Total_Accidents',
    color_continuous_scale=[[0, SKY], [0.5, BLUE], [1, NAVY]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Total Accidents',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

fig.show()
print("✅ Chart 02 done")
```

```python
# Chart 03 — Total Killed by Province
df_killed = df_provinces.groupby('Province')['Killed'].sum().reset_index()
df_killed = df_killed.sort_values('Killed', ascending=True)

fig = px.bar(
    df_killed,
    x='Killed',
    y='Province',
    orientation='h',
    title='💀 Total Fatalities by Province (All Years)',
    color='Killed',
    color_continuous_scale=[[0, '#FCEBEB'], [0.5, RED], [1, DEEP_RED]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Total Killed',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

fig.show()
print("✅ Chart 03 done")
```

```python
# Chart 04 — Average Severity Index by Province
df_severity = df_provinces.groupby('Province')['Severity_Index'].mean().reset_index()
df_severity = df_severity.sort_values('Severity_Index', ascending=True)

fig = px.bar(
    df_severity,
    x='Severity_Index',
    y='Province',
    orientation='h',
    title='⚠️ Average Severity Index by Province',
    color='Severity_Index',
    color_continuous_scale=[[0, SKY], [0.5, RED], [1, DEEP_RED]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Severity Index',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

# Add a vertical reference line at average
avg = df_severity['Severity_Index'].mean()
fig.add_vline(x=avg, line_dash='dash', line_color=NAVY,
              annotation_text=f'Avg: {avg:.2f}', annotation_position='top right')

fig.show()
print("✅ Chart 04 done")
```

```python
# Chart 05 — Fatality Rate Trend Over Years
df_trend_fat = df_national_only.sort_values('Year_Numeric')

fig = px.line(
    df_trend_fat,
    x='Year_Numeric',
    y='Fatality_Rate',
    title='📉 Fatality Rate Trend in Pakistan (2006–2023)',
    markers=True,
    color_discrete_sequence=[RED]
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Year',
    yaxis_title='Fatality Rate (Killed / Accidents)',
    xaxis=dict(tickmode='linear', tick0=2006, dtick=2),
    hovermode='x unified'
)

fig.update_traces(line=dict(width=3), marker=dict(size=8, color=NAVY))

# Add average line
avg_fat = df_trend_fat['Fatality_Rate'].mean()
fig.add_hline(y=avg_fat, line_dash='dash', line_color=BLUE,
              annotation_text=f'Avg: {avg_fat:.2f}',
              annotation_position='top right')

fig.show()
print("✅ Chart 05 done")
```

```python
# Chart 06 — Severity Label Distribution
df_sev_pie = df_provinces['Severity_Label'].value_counts().reset_index()
df_sev_pie.columns = ['Severity_Label', 'Count']

fig = px.pie(
    df_sev_pie,
    names='Severity_Label',
    values='Count',
    title='🥧 Accident Severity Distribution Across Provinces',
    color='Severity_Label',
    color_discrete_map={
        'Low': SKY,
        'Medium': BLUE,
        'High': RED
    },
    hole=0.4
)

fig.update_layout(
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    legend_title='Severity Level',
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    pull=[0.05, 0.05, 0.1]
)

fig.show()
print("✅ Chart 06 done")
```

```python
# Chart 07 — Total Vehicles Involved Over Years
df_vehicles = df_national_only.sort_values('Year_Numeric')

fig = px.area(
    df_vehicles,
    x='Year_Numeric',
    y='Total_Vehicles_Involved',
    title='🚙 Total Vehicles Involved in Accidents (2006–2023)',
    color_discrete_sequence=[BLUE]
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Year',
    yaxis_title='Total Vehicles Involved',
    xaxis=dict(tickmode='linear', tick0=2006, dtick=2),
    hovermode='x unified'
)

fig.update_traces(
    line=dict(width=2, color=NAVY),
    fillcolor='rgba(24, 95, 165, 0.2)'
)

fig.show()
print("✅ Chart 07 done")
```

```python
# Chart 08 — Killed vs Injured by Province
df_ki = df_provinces.groupby('Province')[['Killed', 'Injured']].sum().reset_index()
df_ki = df_ki.sort_values('Killed', ascending=False)

fig = go.Figure()

fig.add_trace(go.Bar(
    name='Killed',
    x=df_ki['Province'],
    y=df_ki['Killed'],
    marker_color=RED
))

fig.add_trace(go.Bar(
    name='Injured',
    x=df_ki['Province'],
    y=df_ki['Injured'],
    marker_color=BLUE
))

fig.update_layout(
    barmode='group',
    title='🏥 Killed vs Injured by Province (All Years)',
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Province',
    yaxis_title='Count',
    legend_title='Category',
    hovermode='x unified'
)

fig.show()
print("✅ Chart 08 done")
```
