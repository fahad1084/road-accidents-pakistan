```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Brand colours
NAVY = '#0C447C'
BLUE = '#185FA5'
SKY = '#378ADD'
RED = '#E24B4A'
DEEP_RED = '#A32D2D'
OFF_WHITE = '#F1EFE8'

# Load datasets
df_main = pd.read_csv('../data/cleaned/cleaned_main.csv')
df_national = pd.read_csv('../data/cleaned/cleaned_national.csv')
df_punjab = pd.read_csv('../data/cleaned/cleaned_punjab.csv')

df_provinces = df_main[df_main['Province'] != 'Pakistan (National)'].copy()
df_national_only = df_main[df_main['Province'] == 'Pakistan (National)'].copy()

print("✅ All variables reloaded successfully")
```

```python
# Chart 09 — Correlation Heatmap
numeric_cols = ['Total_Accidents', 'Fatal_Accidents', 'Non_Fatal_Accidents',
                'Killed', 'Injured', 'Total_Vehicles_Involved',
                'Fatality_Rate', 'Severity_Index']

corr_matrix = df_provinces[numeric_cols].corr().round(2)

fig = px.imshow(
    corr_matrix,
    title='🔥 Correlation Heatmap — Key Accident Features',
    color_continuous_scale=[[0, OFF_WHITE], [0.5, BLUE], [1, NAVY]],
    zmin=-1, zmax=1,
    text_auto=True,
    aspect='auto'
)

fig.update_layout(
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    coloraxis_colorbar=dict(title='Correlation'),
    width=700,
    height=600
)

fig.show()
print("✅ Chart 09 done")
```

```python
# Chart 10 — Fatal vs Non-Fatal Accidents Over Years
df_fatal_trend = df_national_only.sort_values('Year_Numeric')

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_fatal_trend['Year_Numeric'],
    y=df_fatal_trend['Fatal_Accidents'],
    name='Fatal',
    mode='lines+markers',
    line=dict(width=3, color=RED),
    marker=dict(size=7),
    fill='tozeroy',
    fillcolor='rgba(226, 75, 74, 0.2)'
))

fig.add_trace(go.Scatter(
    x=df_fatal_trend['Year_Numeric'],
    y=df_fatal_trend['Non_Fatal_Accidents'],
    name='Non-Fatal',
    mode='lines+markers',
    line=dict(width=3, color=BLUE),
    marker=dict(size=7),
    fill='tozeroy',
    fillcolor='rgba(24, 95, 165, 0.2)'
))

fig.update_layout(
    title='📊 Fatal vs Non-Fatal Accidents Over Years',
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Year',
    yaxis_title='Number of Accidents',
    xaxis=dict(tickmode='linear', tick0=2006, dtick=2),
    hovermode='x unified',
    legend_title='Accident Type'
)

fig.show()
print("✅ Chart 10 done")
```

```python
# Chart 11 — Average Injury Rate by Province
df_injury = df_provinces.groupby('Province')['Injury_Rate'].mean().reset_index()
df_injury = df_injury.sort_values('Injury_Rate', ascending=True)

fig = px.bar(
    df_injury,
    x='Injury_Rate',
    y='Province',
    orientation='h',
    title='🏥 Average Injury Rate by Province',
    color='Injury_Rate',
    color_continuous_scale=[[0, SKY], [0.5, BLUE], [1, NAVY]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Injury Rate (Injured / Accidents)',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

# Add average line
avg_inj = df_injury['Injury_Rate'].mean()
fig.add_vline(x=avg_inj, line_dash='dash', line_color=RED,
              annotation_text=f'Avg: {avg_inj:.2f}',
              annotation_position='top right')

fig.show()
print("✅ Chart 11 done")
```

```python
# Chart 12 — Average Vehicles Per Accident by Province
df_vpa = df_provinces.groupby('Province')['Vehicles_Per_Accident'].mean().reset_index()
df_vpa = df_vpa.sort_values('Vehicles_Per_Accident', ascending=True)

fig = px.bar(
    df_vpa,
    x='Vehicles_Per_Accident',
    y='Province',
    orientation='h',
    title='🚗 Average Vehicles Per Accident by Province',
    color='Vehicles_Per_Accident',
    color_continuous_scale=[[0, SKY], [0.5, BLUE], [1, NAVY]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Vehicles Per Accident',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

# Add average line
avg_vpa = df_vpa['Vehicles_Per_Accident'].mean()
fig.add_vline(x=avg_vpa, line_dash='dash', line_color=RED,
              annotation_text=f'Avg: {avg_vpa:.2f}',
              annotation_position='top right')

fig.show()
print("✅ Chart 12 done")
```

```python
# Chart 13 Final — Top 15 Punjab Districts by Road Accidents
df_pjb_top15 = df_punjab.sort_values('Road_Accidents', ascending=True).tail(15)

fig = px.bar(
    df_pjb_top15,
    x='Road_Accidents',
    y='District',
    orientation='h',
    title='🏙️ Top 15 Punjab Districts by Road Accidents',
    color='Severity_Label',
    color_discrete_map={
        'Low': SKY,
        'Medium': BLUE,
        'High': RED
    }
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Total Road Accidents',
    yaxis_title='District',
    legend_title='Severity',
    hovermode='y unified',
    height=500
)

fig.show()
print("✅ Chart 13 final done")
```

```python
# Chart 14 Final — Top 15 Punjab Districts by Road Accident %
df_pct_top15 = df_punjab.sort_values('Road_Accident_Pct', ascending=True).tail(15)

fig = px.bar(
    df_pct_top15,
    x='Road_Accident_Pct',
    y='District',
    orientation='h',
    title='📊 Top 15 Districts — Road Accidents as % of Emergency Calls',
    color='Road_Accident_Pct',
    color_continuous_scale=[[0, SKY], [0.5, BLUE], [1, NAVY]],
    height=500
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Road Accident % of Total Calls',
    yaxis_title='District',
    coloraxis_showscale=False,
    hovermode='y unified'
)

avg_pct = df_pct_top15['Road_Accident_Pct'].mean()
fig.add_vline(x=avg_pct, line_dash='dash', line_color=RED,
              annotation_text=f'Avg: {avg_pct:.1f}%',
              annotation_position='top right')

fig.show()
print("✅ Chart 14 final done")
```

```python
# Chart 15 — Killed Per Fatal Accident by Province
df_kpf = df_provinces.groupby('Province')['Killed_Per_Fatal_Acc'].mean().reset_index()
df_kpf = df_kpf.sort_values('Killed_Per_Fatal_Acc', ascending=True)

fig = px.bar(
    df_kpf,
    x='Killed_Per_Fatal_Acc',
    y='Province',
    orientation='h',
    title='💀 Average Killed Per Fatal Accident by Province',
    color='Killed_Per_Fatal_Acc',
    color_continuous_scale=[[0, '#FCEBEB'], [0.5, RED], [1, DEEP_RED]],
)

fig.update_layout(
    plot_bgcolor=OFF_WHITE,
    paper_bgcolor='white',
    title_font=dict(size=18, color=NAVY),
    xaxis_title='Killed Per Fatal Accident',
    yaxis_title='Province',
    coloraxis_showscale=False,
    hovermode='y unified'
)

# Add average line
avg_kpf = df_kpf['Killed_Per_Fatal_Acc'].mean()
fig.add_vline(x=avg_kpf, line_dash='dash', line_color=NAVY,
              annotation_text=f'Avg: {avg_kpf:.2f}',
              annotation_position='top right')

fig.show()
print("✅ Chart 15 done")
```
