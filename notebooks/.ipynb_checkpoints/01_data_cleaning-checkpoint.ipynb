```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries loaded successfully")
```

```python
# Load all raw datasets
df_main = pd.read_csv('../data/raw/pakistan_road_accidents_2006_2023.csv')
df_national = pd.read_csv('../data/raw/Traffic Accidents in Pakistan.csv')
df_punjab = pd.read_csv('../data/raw/pakistan_accidents_district_punjab.csv')

print("✅ All datasets loaded")
print(f"Main dataset:     {df_main.shape}")
print(f"National dataset: {df_national.shape}")
print(f"Punjab dataset:   {df_punjab.shape}")
```

```python
# Inspect main dataset
print("=" * 50)
print("MAIN DATASET — Columns & Nulls")
print("=" * 50)
print(df_main.info())

print("\n")
print("=" * 50)
print("NATIONAL DATASET — Columns & Nulls")
print("=" * 50)
print(df_national.info())

print("\n")
print("=" * 50)
print("PUNJAB DATASET — Columns & Nulls")
print("=" * 50)
print(df_punjab.info())
```

```python
# Check what rows are null in national dataset
print("Null rows in national dataset:")
print(df_national[df_national['Total Number of Accident'].isnull()])
```

```python
# Drop the separator rows (they are just province/country name labels)
df_national = df_national[df_national['Total Number of Accident'].notna()].reset_index(drop=True)

# Rename columns to clean names
df_national.columns = [
    'Year', 'Total_Accidents', 'Fatal_Accidents',
    'Non_Fatal_Accidents', 'Killed', 'Injured', 'Total_Vehicles_Involved'
]

print(f"✅ National dataset cleaned: {df_national.shape}")
print(df_national.head())
```

```python
# Extract numeric year from 'Year' column (e.g. '2010-11' -> 2010)
df_main['Year_Clean'] = df_main['Year_Numeric']

# Standardise province names
province_map = {
    'KPK': 'Khyber Pakhtunkhwa',
    'NWFP': 'Khyber Pakhtunkhwa',
    'AJK': 'Azad Jammu & Kashmir',
}
df_main['Province'] = df_main['Province'].replace(province_map)

# Add severity label column based on Severity_Index
def severity_label(val):
    if val < 2.0:
        return 'Low'
    elif val < 2.5:
        return 'Medium'
    else:
        return 'High'

df_main['Severity_Label'] = df_main['Severity_Index'].apply(severity_label)

print("✅ Main dataset standardised")
print(df_main['Province'].unique())
print(df_main['Severity_Label'].value_counts())
```

```python
# Extract year numeric from 'Year' column (e.g. '2012-2013' -> 2012)
df_national['Year_Clean'] = df_national['Year'].str[:4].astype(int)

# Add province column by repeating province names
# The 126 rows = 6 provinces x 21 years
provinces = ['Pakistan (National)', 'Islamabad', 'Punjab', 'Sindh', 'Khyber Pakhtunkhwa', 'Balochistan']
rows_per_province = len(df_national) // len(provinces)
df_national['Province'] = [p for p in provinces for _ in range(rows_per_province)]

# Calculate fatality rate
df_national['Fatality_Rate'] = (df_national['Killed'] / df_national['Total_Accidents']).round(3)

# Add severity label
df_national['Severity_Label'] = df_national['Fatality_Rate'].apply(
    lambda x: 'High' if x > 0.6 else ('Medium' if x > 0.4 else 'Low')
)

print("✅ National dataset feature engineered")
print(df_national['Province'].value_counts())
print(df_national.head(3))
```

```python
# Add road accident severity category based on percentage
df_punjab['Severity_Label'] = df_punjab['Road_Accident_Pct'].apply(
    lambda x: 'High' if x > 35 else ('Medium' if x > 25 else 'Low')
)

# Add accidents per rescue call ratio
df_punjab['Accident_Rescue_Ratio'] = (
    df_punjab['Road_Accidents'] / df_punjab['Total_Calls']
).round(4)

# Clean district names (strip whitespace)
df_punjab['District'] = df_punjab['District'].str.strip()

print("✅ Punjab dataset feature engineered")
print(df_punjab[['District', 'Road_Accidents', 'Road_Accident_Pct', 'Severity_Label']].head(10))
```

```python
import os

# Create cleaned folder if not exists
os.makedirs('../data/cleaned', exist_ok=True)

# Save all 3 cleaned datasets
df_main.to_csv('../data/cleaned/cleaned_main.csv', index=False)
df_national.to_csv('../data/cleaned/cleaned_national.csv', index=False)
df_punjab.to_csv('../data/cleaned/cleaned_punjab.csv', index=False)

print("✅ All cleaned datasets saved!")
print(f"cleaned_main.csv     — {df_main.shape}")
print(f"cleaned_national.csv — {df_national.shape}")
print(f"cleaned_punjab.csv   — {df_punjab.shape}")
```

```python
import openpyxl
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment

# Define data dictionary entries
dictionary = {
    'cleaned_main': [
        ('Year', 'str', 'Fiscal year e.g. 2010-11'),
        ('Province', 'str', 'Province name standardised'),
        ('Total_Accidents', 'int', 'Total road accidents reported'),
        ('Fatal_Accidents', 'int', 'Accidents resulting in death'),
        ('Non_Fatal_Accidents', 'int', 'Accidents with no fatalities'),
        ('Killed', 'int', 'Total persons killed'),
        ('Injured', 'int', 'Total persons injured'),
        ('Total_Vehicles_Involved', 'int', 'Total vehicles involved'),
        ('Source', 'str', 'Data source (NTRC/PBS)'),
        ('Coverage', 'str', 'Geographic coverage'),
        ('Fatality_Rate', 'float', 'Killed / Total_Accidents'),
        ('Injury_Rate', 'float', 'Injured / Total_Accidents'),
        ('Fatal_Accident_Ratio', 'float', 'Fatal_Accidents / Total_Accidents'),
        ('Vehicles_Per_Accident', 'float', 'Vehicles / Total_Accidents'),
        ('Killed_Per_Fatal_Acc', 'float', 'Killed / Fatal_Accidents'),
        ('Injured_Per_Accident', 'float', 'Injured / Total_Accidents'),
        ('Year_Type', 'str', 'Fiscal or Calendar year'),
        ('Year_Numeric', 'int', 'Numeric year for sorting'),
        ('Severity_Index', 'float', 'Composite severity score'),
        ('Province_Code', 'int', 'Numeric province identifier'),
        ('Year_Clean', 'int', 'Cleaned numeric year'),
        ('Severity_Label', 'str', 'Low / Medium / High severity'),
    ],
    'cleaned_national': [
        ('Year', 'str', 'Fiscal year e.g. 2012-2013'),
        ('Total_Accidents', 'float', 'Total accidents nationally'),
        ('Fatal_Accidents', 'float', 'Fatal accidents count'),
        ('Non_Fatal_Accidents', 'float', 'Non-fatal accidents count'),
        ('Killed', 'float', 'Persons killed'),
        ('Injured', 'float', 'Persons injured'),
        ('Total_Vehicles_Involved', 'float', 'Vehicles involved'),
        ('Year_Clean', 'int', 'Numeric year extracted'),
        ('Province', 'str', 'Province or national level'),
        ('Fatality_Rate', 'float', 'Killed / Total_Accidents'),
        ('Severity_Label', 'str', 'Low / Medium / High'),
    ],
    'cleaned_punjab': [
        ('District', 'str', 'Punjab district name'),
        ('Total_Calls', 'int', 'Total Rescue 1122 calls'),
        ('Emergency_Calls', 'int', 'Emergency calls received'),
        ('Road_Accidents', 'int', 'Road accident calls'),
        ('Medical', 'int', 'Medical emergency calls'),
        ('Fire', 'int', 'Fire emergency calls'),
        ('Structure_Collapse', 'int', 'Structure collapse calls'),
        ('Crime_Incidents', 'int', 'Crime incident calls'),
        ('Drowning', 'int', 'Drowning incident calls'),
        ('Misc', 'int', 'Miscellaneous calls'),
        ('Fake_Calls', 'int', 'Fake/prank calls'),
        ('Patients_Rescued', 'int', 'Total patients rescued'),
        ('Province', 'str', 'Province (all Punjab)'),
        ('Source', 'str', 'Data source'),
        ('Coverage', 'str', 'Time coverage'),
        ('Data_Type', 'str', 'Cumulative or annual'),
        ('Road_Accident_Pct', 'float', 'Road accidents as % of total calls'),
        ('Is_RYK', 'bool', 'Is Rahim Yar Khan district'),
        ('Severity_Label', 'str', 'Low / Medium / High'),
        ('Accident_Rescue_Ratio', 'float', 'Road_Accidents / Total_Calls'),
    ],
}

# Create Excel workbook
wb = Workbook()

# Header style
header_font = Font(bold=True, color='FFFFFF')
header_fill = PatternFill(fill_type='solid', fgColor='0C447C')

for sheet_name, columns in dictionary.items():
    ws = wb.create_sheet(title=sheet_name)
    
    # Write headers
    headers = ['Column Name', 'Data Type', 'Description']
    for col_num, header in enumerate(headers, 1):
        cell = ws.cell(row=1, column=col_num, value=header)
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = Alignment(horizontal='center')
    
    # Write rows
    for row_num, (col_name, dtype, desc) in enumerate(columns, 2):
        ws.cell(row=row_num, column=1, value=col_name)
        ws.cell(row=row_num, column=2, value=dtype)
        ws.cell(row=row_num, column=3, value=desc)
        if row_num % 2 == 0:
            for col in range(1, 4):
                ws.cell(row=row_num, column=col).fill = PatternFill(fill_type='solid', fgColor='E6F1FB')
    
    # Set column widths
    ws.column_dimensions['A'].width = 28
    ws.column_dimensions['B'].width = 12
    ws.column_dimensions['C'].width = 45

# Remove default sheet
del wb['Sheet']

# Save
wb.save('../data/cleaned/data_dictionary.xlsx')
print("✅ Data dictionary saved to data/cleaned/data_dictionary.xlsx")
print(f"   Sheets: {wb.sheetnames}")
```
